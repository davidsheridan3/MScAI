My problem at hand:
I am rolling two 6-sided dice and I aim to find out the probability that the sum of the two dice is:
1. Less than 4
2. More than 9


```python
def two_dice_outcome(): #my function to simulate the action and outcomes of rolling 2 dice
    dice_1 = random.randint(1,6)
    dice_2 = random.randint(1,6)
    return dice_1 + dice_2 #outcome is the addition of each outcome value
```

```python
print(two_dice_outcome())
```

```python
trial_num = 100 #number of dice rolls
results = []
```

```python
for _ in range(trial_num):
    result = two_dice_outcome()
    results.append(result)
```

```python
count_less_than_4 = sum(1 for x in results if x < 4)
count_more_than_9 = sum(1 for x in results if x > 9)
count_more_than_4_less_than_9 = sum(1 for x in results if x > 4 and x < 9) #for testing purposes, a different value for probability 

probability_less_than_4 = count_less_than_4 / trial_num
probability_more_than_9 = count_more_than_9 / trial_num
probability_more_than_4_less_than_9 = count_more_than_4_less_than_9 / trial_num
```

```python
print(probability_more_than_9) #test function
```

```python
print(probability_less_than_4) #test function
```

```python
print(probability_more_than_4_less_than_9 #test function
```

```python
mean_result = np.mean(results)
std_dev_result = np.std(results)
```

```python
print(f"Mean of sum: {mean_result}")
```

```python
print(f"Standard Deviation of sum: {std_dev_result}")
```

```python
#Plotting the distribution of results
plt.hist(results, bins=range(2, 14), edgecolor='black', align='left')
plt.title('Distribution of Sum of 2 Dice Rolls')
plt.xlabel('Sum of Dice Rolls')
plt.ylabel('Frequency')
plt.xticks(range(2,13)) 
plt.grid(True)
plt.show() 
```
