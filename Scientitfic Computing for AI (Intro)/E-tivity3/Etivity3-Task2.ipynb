```python
#Tokenisation and Word-Counting Functions
def tokenise(emails):
    words = [] #empty list that will hold all details from emails
    for email in emails:
        words.extend(email.split()) #split email text into individual words
    return words

def build_word_counts(emails):
    word_counts = {} #empty dictionary that will store word counts
    for word in emails:
        word_counts[word.lower()] = word_counts.get(word.lower(),0) + 1 #convert word to lower, if word is already in dictionary increment its count by 1, if not add to dictionary with count of 1
    return word_counts
```

```python
previous_spam = ['send us your password', 'review our website', 'send your password', 'send us your account']
previous_ham = ['Your activity report', 'beneifits physical activity', 'the importance vows']
new_emails = {
    'spam': ['renew your password', 'renew your vows'],
    'ham': ['benefits of our account','the importance of physical activity']
}
```

```python
ham_words = tokenise(previous_ham)
spam_words = tokenise(previous_spam)
```

```python
print(ham_words)
```

```python
ham_word_counts = build_word_counts(ham_words)
spam_word_counts = build_word_counts(spam_words)
```

```python
print(spam_word_counts)
```

```python
total_ham_words = len(ham_words)
total_spam_words = len(spam_words)
```

```python
print(total_ham_words)
```

```python
def word_probability(word, word_counts, total_words, smoothing=1): #probability of word occuring in set of emails

    return(word_counts.get(word.lower(), 0) + smoothing) / (total_words +smoothing * 2)
```

```python
def classify(email, ham_word_counts, spam_word_counts, total_ham_words, total_spam_words, prior_ham, prior_spam): #classifies ham or spam using Naive Bayes
    email_words = email.split()
    
    prob_ham = prior_ham
    prob_spam = prior_spam
    
    for word in email_words: #loops through each word in email to calculate probability ham/spam
        prob_word_given_ham = word_probability(word, ham_word_counts, total_ham_words)
        prob_word_given_spam = word_probability(word, spam_word_counts, total_spam_words)
        
        prob_ham *= prob_word_given_ham
        prob_spam *= prob_word_given_spam
        
    if prob_spam > prob_ham:
        return "spam"
    else:
        return "ham"
```

```python
prior_ham = len(previous_ham) / (len(previous_ham) + len(previous_spam))
prior_spam = len(previous_spam) / (len(previous_ham) + len(previous_spam))
```

```python
for label, emails in new_emails.items():
    for email in emails:
        prediction = classify(email, ham_word_counts, spam_word_counts, total_ham_words, total_spam_words, prior_ham, prior_spam) #classify function to predict ham or spam
        print(f"Email: '{email}' | Actual: {label} | Predicted: {prediction}")
                              
```

```python
def update_classifier(new_emails, ham_word_counts, spam_word_counts): #updates word count for ham/spam
    for label, emails in new_emails.items():
        for email in emails:
            words = email.split()
            if label == "ham":
                for word in words:
                    ham_word_counts[word.lower()] = ham_word_counts.get(word.lower(),0) + 1 #if already in ham dictionary, increment count by 1, if not add initial count by 1
            else:
                for word in words:
                    spam_word_counts[word.lower()] = spam_word_counts.get(word.lower(),0) + 1 #convert to lower and update count for spam
```

```python
update_classifier(new_emails, ham_word_counts, spam_word_counts)
```

```python
print("\n-After Learning-")
for label, emails in new_emails.items():
    for email in emails:
        prediction = classify(email, ham_word_counts, spam_word_counts, total_ham_words, total_spam_words, prior_ham, prior_spam)
        print(f"Email: '{email}' | Actual: {label} | Predicted: {prediction}")
```
