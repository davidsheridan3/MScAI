```python
#Function 1 - To get the derivative of a polynomial
#I used the power rule (coefficient*power and power - 1)


def dr_of_poly(poly_dict): #to calculate the derivative of a polynomial.

    dr_dict = {} #dictionary used (rather than an array) to store results (derivatives).

    for power, coefficient in poly_dict.items(): #to go through each term one by one (coefficient and power)
      if power > 0: #constants don't change when we take the derivative (terms with power greater than 0 only)
        dr_dict[power - 1] = coefficient * power #standard power rule (multiply coefficient by power and reduce power by 1)

    return dr_dict #dictionary containing dervative
```

```python
#Test 1 of Function 1 

#5x^4 - 3x^3 + 7x^2 + 4x - 10
#What to expect using the power rule:
#5x^4 = 20x^3
#-3x^3 = -9x^2
#7x^2 = 14x
#4x = 4
#-10 = 0
#=> we expect an output of 20x^3 - 9x^2 + 14x + 4 {3: 20, 2: -9, 1: 14, 0: 4}

poly_dict = {4: 5, 3: -3, 2: 7, 1: 4, 0: -10} #power: coefficient, power: coefficient.....
derivative = dr_of_poly(poly_dict)
print(derivative) 
```

```python
print(f"The derivative of your input is: {derivative}") #for the users experience 
```

```python
# Function 2 - Polynomial evaluation at a SPECIFIC point

def polynomial_evaluation(poly_dict, x): #polynomial to be given as dictionary alongside a value for 'x'.

    result = 0 #result variable set to 0
    
    for power, coefficient in poly_dict.items(): #power and coefficient seperated for function (one by one)
        
        result += coefficient * (x ** power) #value of x to be raised to it's power, followed being multiplied by the coefficient (if there is one)
        
    return result
```

```python
#Test 1 of Function 2 
#Evaluation of 6x^3 + 8x^2 - 4x + 7 where x = 2
poly_dict = {3: 6, 2: 8, 1: -4, 0: 7} #power: coefficient, power: coefficient.....
x_value = 2 #input
result = polynomial_evaluation(poly_dict, x_value) #this funtion gives correct output
print(result) 
```
